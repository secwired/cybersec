```python
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade
```

```python
# Lista para armazenar os livros cadastrados
livros = []
```

```python
# Função para cadastrar um novo livro
def cadastrar_livro(titulo, autor, genero, quantidade):
    novo_livro = Livro(titulo, autor, genero, quantidade)
    livros.append(novo_livro)
    print(f'Livro "{titulo}" cadastrado com sucesso!')

# Função para listar todos os livros
def listar_livros():
    if not livros:
        print("Nenhum livro cadastrado.")
    else:
        for livro in livros:
            print(f'Título: {livro.titulo}, Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade: {livro.quantidade}')

# Função para buscar um livro pelo título
def buscar_livro(titulo):
    for livro in livros:
        if livro.titulo.lower() == titulo.lower():
            print(f'Livro encontrado: Título: {livro.titulo}, Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade: {livro.quantidade}')
            return
    print("Livro não encontrado.")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Função para gerar gráfico de quantidade de livros por gênero
def gerar_grafico_genero():
    generos = [livro.genero for livro in livros]
    contagem_generos = Counter(generos)

    plt.bar(contagem_generos.keys(), contagem_generos.values())
    plt.title('Quantidade de Livros por Gênero')
    plt.xlabel('Gêneros')
    plt.ylabel('Quantidade de Livros')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Exemplo de uso do sistema
cadastrar_livro("O Senhor dos Anéis", "J.R.R. Tolkien", "Fantasia", 5)
cadastrar_livro("1984", "George Orwell", "Ficção", 3)
cadastrar_livro("A Revolução dos Bichos", "George Orwell", "Ficção", 2)

listar_livros()
buscar_livro("1984")
gerar_grafico_genero()
```
